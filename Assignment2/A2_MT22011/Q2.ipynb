```python
# importing libraries
import os
import random
from  PIL import Image,ImageEnhance,ImageOps
import matplotlib.pyplot as plt
```

```python
# loading images directory
images = os.listdir(os.path.expanduser('~/Desktop/cancer_image'))
```

```python
# loading 3 image randomly
imgageName = [str(k)+".bmp" for k in [random.randint(101,110) for i in range(3)]]
```

```python
imagePaths = [os.path.join('Desktop/cancer_image',imgageName) 
              for imgageName in imgageName] # storing path of each image

fig,ax = plt.subplots(1, 3)
c=0

# visualizing each image along with size
for img in imagePaths:
    img = Image.open(img)
    ax[c].imshow(img)
    c=c+1
    print(img.size)
```

### Resize

```python
c=0
fig,ax = plt.subplots(1,3)
for img in imagePaths:
    img = Image.open(img)
    resizeImg = img.resize((600,900))
    ax[c].imshow(resizeImg)
    c+=1  
```

### Pad

```python
c = 0
fig,ax = plt.subplots(1,3)
for img in imagePaths:
    img = Image.open(img)
    up = 200
    down = 50
    left = 150
    right = 100
    width,height = img.size
    newHeight = height+up+down
    newWidth = width+left+right
    imgNew = Image.new(img.mode,(newWidth,newHeight),(0,0,255))
    imgNew.paste(img,(left,up))
    ax[c].imshow(imgNew)    
    c=c+1
```

### Crop

```python
c = 0
fig,ax = plt.subplots(1,3)
for img in imagePaths:
    img = Image.open(img)
    co_ord = (1650, 1020, 2000, 1980)
    croppedImage = img.crop(co_ord)
    ax[c].imshow(croppedImage)
    c+=1 
```

### Gray Scale

```python
c = 0
fig,ax = plt.subplots(1,3)
for img in imagePaths:
    img = Image.open(img)
    m = ImageOps.grayscale(img)
    ax[c].imshow(m)
    c = c+1   
```

### Contrast

```python
c = 0
fig,ax = plt.subplots(1,3)
for img in imagePaths:
    img = Image.open(img)
    contrast = ImageEnhance.Contrast(img)
    contrastImg = contrast.enhance(5.5)
    ax[c].imshow(contrastImg)
    c = c+1
```

### Saturation

```python
c = 0
fig,ax = plt.subplots(1,3)
for img in imagePaths:
    img = Image.open(img)
    filterImg = ImageEnhance.Color(img)
    satImage = filterImg.enhance(4.5)
    ax[c].imshow(satImage)    
    c = c+1
```
